```python
from sklearn.datasets import load_boston
import pandas as pd
Boston_Data=load_boston()
Boston_X=pd.DataFrame(data=Boston_Data.data,columns=Boston_Data.feature_names)
Boston_Y=pd.DataFrame(Boston_Data.target,columns=['medv'])
Boston_DataFrame=pd.concat([Boston_X,Boston_Y],axis=1)
Boston_DataFrame.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler_Boston=MinMaxScaler().fit(Boston_X)
pd.DataFrame(Scaler_Boston.transform(Boston_X),columns=Boston_Data.feature_names)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

Split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=0)
for Train_index,Test_index in Split.split(Boston_Data,Boston_Data['medv']):
  Train_Boston=Boston_Data.loc[Train_index]
  Test_Boston=Boston_Data.loc[Test_index]
  Train_Boston_x,Train_Boston_y=Train_Boston.iloc[:,:-1],Train_Boston['medv']
  Test_Boston_x,Test_Boston_y=Test_Boston.iloc[:,:-1],Test_Boston['medv']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

Scaler_Boston=MinMaxScaler()
KNN_CLF=KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=2,weights='uniform')
Full_pipeline_KNN=Pipeline(steps=[('scaler',Scaler_Boston),('Classifier',KNN_CLF)])
Full_pipeline_KNN.fit(Train_Boston_x,Train_Boston_y)
Test_Boston_Pred=Full_pipeline_KNN.predict(Test_Boston_x)
pd.DataFrame(confusion_matrix(Test_Boston_y,Test_Boston_Pred))
print("Accuracy=",Full_pipeline_KNN.score(Test_Boston_x,Test_Boston_y))
```
